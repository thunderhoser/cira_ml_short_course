# Import common packages

The next cell imports commonly used Python packages.

```
%matplotlib inline
import numpy
from matplotlib import pyplot

FONT_SIZE = 20
FIGURE_WIDTH_INCHES = 10
FIGURE_HEIGHT_INCHES = 10

pyplot.rc('font', size=FONT_SIZE)
pyplot.rc('axes', titlesize=FONT_SIZE)
pyplot.rc('axes', labelsize=FONT_SIZE)
pyplot.rc('xtick', labelsize=FONT_SIZE)
pyplot.rc('ytick', labelsize=FONT_SIZE)
pyplot.rc('legend', fontsize=FONT_SIZE)
pyplot.rc('figure', titlesize=FONT_SIZE)
```

# Do simple math

The next cell solves a simple math problem.

```
side_lengths = numpy.array([3, 4], dtype=float)
hypotenuse_length = numpy.sqrt(numpy.sum(side_lengths ** 2))

print('Hypotenuse length = {0:f}'.format(hypotenuse_length))
```

# Create plot

The next cell creates a simple plot.  Typing `%matplotlib inline` above is what allows this plot to be displayed directly in the notebook.

```
gaussian_values = numpy.random.normal(loc=0., scale=1., size=int(1e7))

figure_object, axes_object = pyplot.subplots(
    1, 1, figsize=(FIGURE_WIDTH_INCHES, FIGURE_HEIGHT_INCHES)
)

axes_object.hist(gaussian_values, bins=20)
axes_object.set_title('Standard normal distribution')
axes_object.set_xlabel('Value')
axes_object.set_ylabel('Frequency')
pyplot.show()
```
